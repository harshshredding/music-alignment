```python
import csv
import subprocess
import os

def splitMidiFile(filepath):
#     filepath = 'bach_wtc/MIDI-Unprocessed_050_PIANO050_MID--AUDIO-split_07-06-17_Piano-e_3-01_wav--1.midi'
    filename = os.path.split(filepath)[1]
    filename_without_extension, _ = os.path.splitext(filename)
    csv_full_file_name = 'work/' + filename_without_extension + '.csv'
    csv_prelude_file_name = 'work/' + filename_without_extension + '_prelude.csv'
    csv_fugue_file_name = 'work/' + filename_without_extension + '_fugue.csv'
    midi_full_file_name = 'work/' + filename_without_extension + '.midi'
    midi_prelude_file_name = 'work/' + filename_without_extension + '_prelude.midi'
    midi_fugue_file_name = 'work/' + filename_without_extension + '_fugue.midi'


    print(filename_without_extension)
    print(csv_full_file_name)
    print(csv_prelude_file_name)
    print(csv_fugue_file_name)
    print(midi_full_file_name)
    print(midi_prelude_file_name)
    print(midi_fugue_file_name)

    subprocess.run(["../midicsv-1.1/midicsv", filepath, csv_full_file_name])
    audio_split, csv_split = musiclib.findSeperationPoint(filepath)
    with open(csv_full_file_name, newline='') as inputFile, open(csv_prelude_file_name, 'w', newline='') as preludeFile, open(csv_fugue_file_name, 'w', newline='') as fugueFile:
        reader = csv.reader(inputFile)
        writer_prelude = csv.writer(preludeFile)
        writer_fugue = csv.writer(fugueFile)
        count_till_fugue_start = 0
        time_to_subtract = 0
        reached_start_of_fugue = False
        for row in reader:
            if row[0] == '2':
                count_till_fugue_start = count_till_fugue_start + 1
                if count_till_fugue_start == csv_split:
                    reached_start_of_fugue = True
                    time_to_subtract = row[1]
                    prelude_end_row = [row[0], row[1], 'End_track']
                    fugue_start_row = ['2', '0', 'Start_track']
                    writer_prelude.writerow(prelude_end_row)
                    writer_fugue.writerow(fugue_start_row)
                if reached_start_of_fugue:
                    row[1] = int(row[1]) - int(time_to_subtract)
                    writer_fugue.writerow(row)
                else:
                    writer_prelude.writerow(row)
            else:
                writer_prelude.writerow(row)
                writer_fugue.writerow(row)

    subprocess.run(["../midicsv-1.1/csvmidi", csv_full_file_name, midi_full_file_name])
    subprocess.run(["../midicsv-1.1/csvmidi", csv_prelude_file_name, midi_prelude_file_name]) 
    subprocess.run(["../midicsv-1.1/csvmidi", csv_fugue_file_name, midi_fugue_file_name]) 
```
